```python
# mengimport library 
import numpy as np #lib numpy
import imageio     #lib imageio
import matplotlib.pyplot as plt  #lib matplotlib
```

```python
img = imageio.imread("ASIMO.jpg") #memanggil/membaca gambar
```

```python
#resolusi dan tipe gambar
img_height = img.shape[0] #tinggi gambar
img_width = img.shape[1]  #Lebar gambar
img_channel = img.shape[2] #channel gambar
img_type = img.dtype     #tipe gambar
```

```python
img_flip_horizontal = np.zeros(img.shape, img_type) #Membuat variabel horizontal dengan resolusi dan tipe yang sama seperti gambar
img_flip_vertical = np.zeros(img.shape, img_type) #Membuat variabel vertikal dengan resolusi dan tipe yang sama seperti gambar
```

```python
for y in range(0, img_height): # y gambar
    for x in range(0, img_width): # x gambar
        for c in range(0, img_channel): #channel gambar
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] #Membalik gambar secara horizontal
```

```python
for y in range(0, img_height): # y gambar
    for x in range(0, img_width): # x gambar 
        for c in range(0, img_channel): #channel gambar
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c] #Membalik gambar secara vertikal
```

```python
plt.imshow(img_flip_horizontal)  #membalikan gambar secara horizontal
plt.title("Flip Horizontal") #judul 
plt.show() #tampilkan gambar
plt.imshow(img_flip_vertical) #membalikan gambar secara vertikal
plt.title("Flip Vertical") #judul 
plt.show() #tampilkan gambar
```
